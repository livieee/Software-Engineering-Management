# Hypothesis Testing
## Hypothesis:
- Neuroprotective microglial activation during Mild Cognitive Impairment (MCI) stages.
- Neurotoxic microglial activation during later AD stages with tau pathology.
- Exhibit heightened inflammatory responses, and correlates with disease progression.

Step 1: Defines the MCI and AD groups based on Braak stage, ADNC, and cognitive status.

Step 2: Extracts gene expression data specifically for MCI and AD cells.

Step 3: Implements a function to compute the Mann-Whitney U test (Wilcoxon rank-sum test) to compare expression levels between the two groups.

Step 4: Analyzes a set of neuroprotective (TREM2, CD68) and pro-inflammatory (IL1B, TNF) genes of interest between ADSN Scores and stages.



```python
import pandas as pd
import numpy as np
from scipy.stats import mannwhitneyu, wilcoxon

# Step 1: Ensure the MCI and AD groups are defined properly
adata.obs['MCI'] = (
    (adata.obs['Braak stage'].isin(['Braak II', 'Braak III'])) &
    (adata.obs['ADNC'].isin(['No AD', 'Low'])) &
    (adata.obs['Cognitive status'] == 'No dementia')
).astype('category')

adata.obs['Alzheimers_Disease'] = (
    (adata.obs['Braak stage'].isin(['Braak V', 'Braak VI'])) &
    (adata.obs['ADNC'].isin(['Intermediate', 'High'])) &
    (adata.obs['Cognitive status'] == 'Dementia')
).astype('category')

# Step 2: Extract gene expression data for MCI and AD groups
mci_cells = adata[adata.obs['MCI'] == True]
ad_cells = adata[adata.obs['Alzheimers_Disease'] == True]

# Step 3: Define a function to compute statistical tests for gene expression differences
def differential_expression_test(gene_id):
    """Perform statistical test (Wilcoxon rank-sum) to compare MCI vs AD."""
    mci_expr = mci_cells[:, gene_id].X.toarray().flatten()  # Convert sparse to dense
    ad_expr = ad_cells[:, gene_id].X.toarray().flatten()

    # Perform Wilcoxon rank-sum test
    stat, p_value = mannwhitneyu(mci_expr, ad_expr, alternative='two-sided')
    return stat, p_value

# Step 4: Analyze differences for neuroprotective and pro-inflammatory genes
genes_of_interest = ['ENSG00000095970', 'ENSG00000129226',  # TREM2, CD68
                     'ENSG00000125538', 'ENSG00000232810']  # IL1B, TNF

results = []
for gene in genes_of_interest:
    stat, p_value = differential_expression_test(gene)
    results.append({'Gene': gene, 'Stat': stat, 'p-value': p_value})

# Step 5: Display the results in a DataFrame
results_df = pd.DataFrame(results)
print("Differential Expression Results between MCI and AD Groups:")
print(results_df)

# Step 6: Plot gene expression differences for visualization
import seaborn as sns
import matplotlib.pyplot as plt

for gene in genes_of_interest:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=adata.obs, x='MCI', y=adata[:, gene].X.toarray().flatten(), palette='muted')
    sns.stripplot(data=adata.obs, x='MCI', y=adata[:, gene].X.toarray().flatten(), color='black', alpha=0.3)
    plt.title(f"Expression of {gene} in MCI vs AD Groups")
    plt.show()
```


## Validation of Hypothesis - Dual Effect
The plots support the idea that neuroprotective genes (TREM2 and CD68) are more expressed during MCI, facilitating microglial activation and amyloid clearance.

Pro-inflammatory markers (IL1B and TNF) are more expressed in later stages, such as high level ADNC socre, indicating the transition to neurotoxicity and inflammation.
