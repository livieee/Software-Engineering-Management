**Insights:**

Both plots suggest no strong evidence that PMI differs significantly between Alzheimer’s and normal groups. However, some slight shifts in the distributions are visible.

Alzheimer patients tend to have slightly lower PMI values on average (seen from the KDE peaks), while normal individuals exhibit a higher density in the 20-25 PMI range.

Since PMI measures the time between death and sample preservation, understanding whether these shifts are statistically significant could provide insights into how post-mortem delays affect measurements, but from these visualizations alone, the difference appears subtle.

```python
from scipy.stats import mannwhitneyu

# Ensure PMI is numeric
adata.obs['PMI'] = adata.obs['PMI'].astype(float)

# Extract PMI values for both groups directly from adata.obs
pmi_alzheimer = adata.obs[adata.obs['disease'] == 'Alzheimer disease']['PMI']
pmi_normal = adata.obs[adata.obs['disease'] == 'normal']['PMI']

# Perform Mann-Whitney U test to compare the two distributions
stat, p_value = mannwhitneyu(pmi_alzheimer, pmi_normal, alternative='two-sided')

# Display the results
print(f"Mann-Whitney U Test Statistic: {stat}")
print(f"P-value: {p_value}")

# Interpretation based on p-value
if p_value < 0.05:
    print("There is a statistically significant difference in PMI between the two groups.")
else:
    print("There is no statistically significant difference in PMI between the two groups.")
```

Since the p-value is much smaller than 0.05, we reject the null hypothesis that there is no difference between the Alzheimer’s and normal groups in terms of PMI.

This suggests that there is a statistically significant difference in the PMI (Post-Mortem Interval) between the two groups.

**Practical Interpretation:**
Although the boxplot and KDE plot showed some overlap between the PMI distributions for the Alzheimer’s disease and normal groups, this test confirms that the difference is statistically significant.
This indicates that PMI might play a role or correlate with the diagnosis status, although further investigation (such as causal analysis) would be needed to confirm the nature and importance of this relationship.

```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_ind

# Extract Alzheimer and Normal group masks
alzheimer_mask = adata.obs['disease'] == 'Alzheimer disease'
normal_mask = adata.obs['disease'] == 'normal'

# Efficiently extract data from sparse matrices without converting them to dense arrays
alzheimer_data = adata[alzheimer_mask].X
normal_data = adata[normal_mask].X

# Calculate mean expression across genes directly on sparse matrices
alzheimer_mean = np.array(alzheimer_data.mean(axis=0)).flatten()
normal_mean = np.array(normal_data.mean(axis=0)).flatten()

# Perform t-test using sparse matrix slices (convert slices to arrays on the fly)
t_stats, p_values = ttest_ind(
    alzheimer_data.A, normal_data.A, axis=0, equal_var=False
)

# Create DataFrame with results
gene_diff_df = pd.DataFrame({
    'Gene_ID': adata.var.index,
    'Gene_Name': adata.var['feature_name'].values,
    'Alzheimer_Expression': alzheimer_mean,
    'Normal_Expression': normal_mean,
    'Expression_Difference': alzheimer_mean - normal_mean,
    'p_value': p_values
})

# Add 'Significant' column and sort by expression difference
gene_diff_df['Significant'] = gene_diff_df['p_value'] < 0.05
gene_diff_df_sorted = gene_diff_df.sort_values(by='Expression_Difference', ascending=False)

# Display the top 10 differentially expressed genes
print("Top 10 Differentially Expressed Genes (Alzheimer - Normal):")
print(gene_diff_df_sorted.head(10)[['Gene_ID', 'Gene_Name', 'Expression_Difference', 'p_value']])
```

```python
def get_top_genes(adata_group, n_top=20):
    """Extract top n expressed genes with gene IDs and names."""
    # Extract expression matrix and calculate mean expression
    expression_matrix = adata_group.X.toarray() if hasattr(adata_group.X, "toarray") else adata_group.X
    mean_expression = expression_matrix.mean(axis=0).flatten()

    # Create DataFrame with gene IDs, gene names, and mean expression
    top_genes = pd.DataFrame({
        'Gene_ID': adata_group.var.index,  # Ensembl Gene IDs
        'Gene_Name': adata_group.var['feature_name'].values,  # Gene names
        'Mean_Expression': mean_expression
    }).sort_values(by='Mean_Expression', ascending=False).head(n_top)

    return top_genes

# Step 3: Extract top 20 genes for each group
top_genes_alz = get_top_genes(alz, n_top=20)
top_genes_normal = get_top_genes(normal, n_top=20)

# Step 4: Display the top genes
print("\nTop 20 Expressed Genes in Alzheimer's Disease:")
print(top_genes_alz)

print("\nTop 20 Expressed Genes in Normal Group:")
print(top_genes_normal)
```

To determine if there is a statistically significant difference between genders in terms of disease prevalence (e.g., Alzheimer's disease vs. normal), you can use a Chi-square test for independence. This test assesses whether there is a significant association between two categorical variables: gender and disease status.

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Step 1: Create a contingency table
contingency_table = pd.crosstab(adata.obs['sex'], adata.obs['disease'])

# Display the contingency table
print("\nContingency Table (Gender vs Disease):")
print(contingency_table)

# Step 2: Perform the Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Step 3: Display the results
print(f"\nChi-square Statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of Freedom: {dof}")
print("\nExpected Frequencies:")
print(pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns))

# Step 4: Interpret the results
alpha = 0.05  # significance level
if p < alpha:
    print("\nThere is a statistically significant difference in disease distribution between genders.")
else:
    print("\nThere is no statistically significant difference in disease distribution between genders.")
```

Peform age-corrected analysis and determine if the higher Alzheimer's disease prevalence in females is truly gender-based or simply reflects an artifact of age distribution, we can apply logistic regression. This will allow us to model the relationship between disease status (dependent variable) and both age and gender (independent variables) to control for the effect of age.

```python
# Ensure all necessary libraries are imported
import pandas as pd
import numpy as np  # Import NumPy to handle mathematical operations
import statsmodels.api as sm

# Create a binary variable: 1 for Alzheimer's disease, 0 for normal
adata.obs['is_alzheimer'] = adata.obs['disease'].apply(lambda x: 1 if x == 'Alzheimer disease' else 0)

# Extract relevant features: 'is_alzheimer' (dependent), 'age', 'sex' (independent)
data = adata.obs[['is_alzheimer', 'age', 'sex']].dropna()

# Convert 'sex' into a binary variable: 1 for female, 0 for male
data['sex_binary'] = data['sex'].apply(lambda x: 1 if x == 'female' else 0)

# Define the dependent and independent variables
X = data[['age', 'sex_binary']]  # Independent variables (age and sex)
X = sm.add_constant(X)  # Add intercept for logistic regression
y = data['is_alzheimer']  # Dependent variable (Alzheimer's status)

# Fit the logistic regression model
logit_model = sm.Logit(y, X).fit()

# Display the logistic regression summary
print(logit_model.summary())

# Extract key results: Odds Ratios and Confidence Intervals
odds_ratios = pd.DataFrame(
    {"Odds Ratio": logit_model.params.apply(lambda x: np.exp(x)),
     "Lower CI": logit_model.conf_int()[0].apply(lambda x: np.exp(x)),
     "Upper CI": logit_model.conf_int()[1].apply(lambda x: np.exp(x))}
)

print("\nOdds Ratios with 95% Confidence Intervals:")
print(odds_ratios)

# Analyze the significance of 'sex' after controlling for 'age'
if logit_model.pvalues['sex_binary'] < 0.05:
    print("\nThere is a statistically significant gender effect on Alzheimer's disease prevalence, even after controlling for age.")
else:
    print("\nThere is no statistically significant gender effect on Alzheimer's disease prevalence after controlling for age.")
```

- Sex (Female vs. Male) (Odds Ratio: 3.544): Females are 3.54 times more likely to have Alzheimer’s disease than males, even after controlling for age. The p-value (< 0.05) suggests this gender difference is statistically significant.

  This finding supports the conclusion that the higher prevalence of Alzheimer's disease among females is not just an artifact of age distribution but reflects a genuine gender difference.

## Differential Gene Expression of Interest Analysis by Gender


```python
import scanpy as sc
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define genes of interest
genes_of_interest = [
    'SLC26A3', 'RASGEF1B', 'LINGO1', 'PDE4DIP', 'LINC01609', 'PHYHIP'
]

# Step 1: Create a mapping of gene names to gene IDs
if 'feature_name' in adata.var.columns:
    gene_mapping = dict(zip(adata.var['feature_name'], adata.var.index))
else:
    raise ValueError("The dataset does not contain 'feature_name' in adata.var.")

# Step 2: Match genes of interest with dataset gene IDs
matched_genes = {gene: gene_mapping[gene] for gene in genes_of_interest if gene in gene_mapping}

# Step 3: Filter the dataset using matched gene IDs
filtered_data = adata[:, list(matched_genes.values())]

# Step 4: Calculate mean expression by disease and gender
disease_groups = ['Alzheimer disease', 'normal']
genders = ['female', 'male']

# Prepare a DataFrame to store results
results = []

for disease in disease_groups:
    for gender in genders:
        mask = (adata.obs['disease'] == disease) & (adata.obs['sex'] == gender)
        avg_expression = np.array(filtered_data[mask].X.mean(axis=0)).flatten()
        for i, gene_name in enumerate(matched_genes.keys()):
            results.append({
                'Gene': gene_name,
                'Disease': disease,
                'Gender': gender,
                'Average_Expression': avg_expression[i]
            })

# Convert results to DataFrame
df_results = pd.DataFrame(results)

# Step 5: Calculate the expression difference between disease and normal groups for each gender
df_pivot = df_results.pivot_table(
    index=['Gene', 'Gender'],
    columns='Disease',
    values='Average_Expression'
).reset_index()

df_pivot['Expression_Difference'] = df_pivot['Alzheimer disease'] - df_pivot['normal']

# Step 6: Visualize the expression differences by gender
plt.figure(figsize=(14, 8))
sns.barplot(
    x='Expression_Difference', y='Gene', hue='Gender',
    data=df_pivot, palette=sns.color_palette(['#8B5FBF', '#7AC47A'])  # Seaborn purple and green colors
)

plt.title('Gene Expression Difference: Alzheimer vs Normal, by Gender')
plt.xlabel('Expression Difference (Alzheimer - Normal)')
plt.ylabel('Gene')
plt.legend(title='Gender', loc='upper right')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
